## Exp 1

```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import mark_inset
import pickle
import numpy as np

with open('./results/exp1_landmarks_mnist.pkl', 'rb') as f:
    data_ = pickle.load(f)

f, ax = plt.subplots(1, 1, figsize=(6, 4))
# plt.ylim((0.0225, 0.04))
plt.plot(data_['landmarks'], np.mean(data_['error']*100, axis=2).T,linewidth=2)
plt.xlabel('The number of Nystroem landmarks $M$')
plt.ylabel('Error rate (%)')
plt.legend(data_['name'])
plt.title('MNIST, $m=60$')

plt.show()

f.savefig('./results/exp1_landmarks_mnist.pdf', bbox_inches='tight')
```
